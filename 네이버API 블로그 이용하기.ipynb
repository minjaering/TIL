우리는 이제까지 인터넷 익스플로러, 크롬과 같은 브라우저로 네이버에 접속한 후, 네이버에서 제공하는 여러 기능과 콘텐츠를 이용했다.  
하지만 파이썬으로 네이버OpenAPI를 프로토콜에 따라 호출하면 브라우저로 일일이 접속하지 않아도 된다. 한 번 파이썬으로 코딩해 놓으면  
블로그, 뉴스 정보를 불러올 수 있고 보고서 작성까지 자동화할 수 있다. 

# HTTP(Hypertext Transfer Protocol)란?
프로토콜(HTTP)이란 인터넷 상에서 클라이언트와 서버가 데이터를 송수신하기 위해 만들어진 규칙이다.  
네이버OpenAPI도 프로토콜에 따라 작동한다. 

클라이언트 :  서비스를 요청(request)하는 측  
\- 서비스를 요청하는 사람, 컴퓨터 혹은 인터넷 브라우저 등 서비스를 요청하는 행위에 대하여 포괄적인 의미로 사용된다.  

서버 : 클라이언트의 요청에 응답(response)하여 서비스를 제공하는 컴퓨터  

메서드(method) : 요청(request)의 방식
\- GET(정보 조회), POST(정보 제출), PUT(정보 수정), DELETE(정보 삭제)  
\- 이 중 블로그, 뉴스 데이터를 가져오는데 사용되는 네이버OpenAPI는 GET 방식이다.  

그래서 우리는 네이버OpenAPI를 이용할 때 프로토콜의 규칙에 따라 코딩을 진행하면 된다.  
블로그, 뉴스 데이터를 GET 방식으로 요청(request)하는 코드를 작성하여 네이버OpenAPI 서버에 보내면  
서버가 응답(response)하여 되돌려준 데이터를 받을 수 있다.

# 블로그 검색 API 살펴보기

1. GET이란 해당 API가 서버에 데이터를 조회하는 작업을 수행한다는 의미이다. 
2. 이러한 작업을 진행하기 위하는 표기된 요청 URL을 사용하여 데이터를 요청해야 한다. 
3. 서버에서 클라이언트(요청자)의 요청에 응답하여 데이터를 전달할 때에는 XML혹은 JSON 형식으로 전달해 준다.

요청변수를 별도로 입력하지 않았다면 해당 요청변수는 위에 표기된 대로 기본값으로 설정된다.  
요청변수를 입력할 때에는 요청URL의 뒤에 ?를 붙인후 입력하면 되고, 요청변수가 여러개라면 &를 붙여 구별할 수 있다.  
즉 요청URL?요청변수=값&요청변수=값 …의 형태로 전달하면 된다.

# 네이버 블로그 검색 API 예제

```python
#API 사용을 위한 os, sys, urllib.request 모듈을 import키워드를 사용하여 불러온다.
import os
import sys
import urllib.request
```

```python
client_id = "아이디"
client_secret = "비밀번호"
```

```python
#퍼센트 인코딩 : 기본적으로 URL에는 한글입력이 제한되는데, 
#대다수의 네이버 사용자는 검색 키워드를 한글로 작성하기 때문에 
#quote함수를 사용하여 퍼센트 인코딩을 진행하는 과정이다.
encText = urllib.parse.quote("조거팬츠") #검색할단어
```

```python
encText
```

```python
url = "https://openapi.naver.com/v1/search/blog?query=" + encText # json 결과 - 도큐먼트에 쿼리 나와있다.
```

```python
#요청URL과 검색할 단어를 합쳐 서버로 전송할 전체 URL을 만든다. 
#이때 API 인증정보가 담긴 header도 함께 보낸다.
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
```

```python
# urlopen함수는 서버에 정보를 요청하고 서버가 돌려준 응답을 받는 역할을 한다. 
# request변수를 urlopen함수의 매개변수로 입력하여 서버에 정보를 요청하면, 서버가 이에 응답하여 전달한 정보는 response 변수에 할당된다
response = urllib.request.urlopen(request)
```

```python
# 서버에서 응답한 정보 중 HTTP 상태코드(HTTP status code)를 rescode에 할당한다. 
# HTTP 상태코드는 요청이 정상적으로 수행되었는지 여부를 알려준다.
rescode = response.getcode()
```

```python
# HTTP 상태코드가 200이면 정상적으로 요청이 진행된 경우이다. 
# 이 때 서버에서 보낸 정보를 읽고, 이를 utf-8로 디코드하여 출력한다. 
# HTTP 상태코드가 200이 아닐경우 에러코드를 출력한다.
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```

# 블로그 현황 일일 보고서 만들기 

```python
#json 형식을 처리하기 위한 json 모듈을 불러온다. 
#json이란 최근 각광받고 있는 데이터교환형식 중 하나다.
import json
data=json.loads(response_body) # loads : 제이슨 모듈을 가져와서 파이썬에서 사용하기위해 사용
```

#### 판다스의 데이터프레임


```python
#데이터프레임은 데이터, 인덱스, 칼럼을 차례대로 입력하여 만들 수 있다.
##빈데이터 프레임을 먼저 만든후에 데이터를 넣는 경우도 많다.
df=DataFrame(None,columns=data['items'][0].keys(),index=range(1,11)) #10개의 데이터가 기본값이으므로 10개만 받는다.
df
```

```python
#뉴스 검색결과의 딕셔너리에서 공통 key를 추출한다.
cols=list(data['items'][0].keys())
```

```python
cols
```

```python
#중첩 for문으로 네이버OpenAPI에서 가져온 데이터를 새로 만든 데이터프레임에 입력한다.
for x in range(0,10):
    for y in range(0,6):
        df.iloc[x,y]=data['items'][x][cols[y]]
df
```

```python
#데이터프레임의 칼럼순서를 엑셀 보고서 형식에 맞추어 변경한다..-> 칼럼순서 바꾸기
df=df[['title', 'description', 'link', 'postdate', 'bloggername', 'bloggerlink']] 
df
```

```python
df['postdate']
```

```python
#날짜데이터의 타입이 문자열이기 때문에 숫자형으로 변환한다.
df['postdate']=df['postdate'].astype(int)
```

```python
#정규표현식을 사용하여 html태그를 삭제한다.
import re
tag=re.compile('<.*?>')
df['title']=df['title'].str.replace(tag,'')
df['description']=df['description'].str.replace(tag,'')
```
